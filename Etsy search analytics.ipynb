```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('Etsy_analystics.xlsx', names=['Search query', 'Impressions', 'Position', 'Listings'])
```

```python
df
```

```python
#Data cleaning
df['Search query'] = df['Search query'].str.strip('a')
```

```python
df['Search query'] = df['Search query'].str.strip('Q')
df['Search query'] = df['Search query'].str.strip('A')
df['Search query'] = df['Search query'].str.strip('7')
df['Search query'] = df['Search query'].str.strip('z')
df['Search query'] = df['Search query'].str.strip('x')
df['Search query'] = df['Search query'].str.strip('2')
df['Search query'] = df['Search query'].str.strip('O')
```

```python
df
```

```python
#Listing = product Listings that appeared for this search query
#Position = Avg position of your listings when appearing in a search
#Impression = Total appearances in search

df_max_impression = df[df['Impressions'] == df['Impressions'].max()]

df_min_position = df[df['Position'] == df['Position'].min()]

df_max_listing = df[df['Listings'] == df['Listings'].max()]

```

```python
display(df_max_impression)
print('The max impression is ', df_max_impression['Impressions'])
```

```python
display(df_min_position)
print('The min position is ', df_min_position['Position'])
```

```python
display(df_max_listing)
print('The max listing is ', df_max_listing['Listings'])
```

```python
top_search_query = df[(df['Impressions']>10)|(df['Position']<35)]
top_search_query.reset_index()
display(top_search_query)
```
